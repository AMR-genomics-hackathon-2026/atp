```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import data
raw_data = pd.read_csv('contig_lengths.txt', sep=' ', header=None, names=['Contig', 'Length'])

# remove "len=" from the Length column and convert to numeric
raw_data['Length'] = raw_data['Length'].str.replace('len=', '').astype(int)

# plot histogram of contig lengths, with kde overlay
plt.figure(figsize=(10, 6))
sns.histplot(raw_data['Length'], bins=50, kde=True, color='blue')
plt.xlabel('Contig Length (bp)')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
# save plot as svg
plt.savefig('contig_length_histogram.svg', format='svg')

# print max contig length
max_length = raw_data['Length'].max()
print(f'Max Contig Length: {max_length} bp')

# print max contig length contig name
max_length_contig = raw_data.loc[raw_data['Length'].idxmax(), 'Contig']
print(f'Contig with Max Length: {max_length_contig}')
```
